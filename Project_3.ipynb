```python
total = len(df)
age_17 = len(df[df['age'] == 17]) / total
age_18 = len(df[df['age'] == 18]) / total
print(f"(17 ∪ 18) = {age_17 + age_18:.3f}")
```

```python
male_urban = len(df[(df['sex'] == 'M') & (df['address'] == 'U')]) / total
print(f"Male ∩ Urban = {male_urban:.3f}")
```

```python
G3_lt_10 = len(df[df['G3'] < 10]) / total
print(f"G3 >= 10 = 1 - {G3_lt_10:.3f} = {1 - G3_lt_10:.3f}")
```

```python
no_fail = df[df['failures'] == 0]
if len(no_fail) > 0:
    cond_prob = len(no_fail[no_fail['G3'] >= 15]) / len(no_fail)
    print(f"4. Conditional: P(G3>=15 | failures=0) = {cond_prob:.3f}")
else:
    print("4. No students with failures=0")
```

```python
high_G3 = df[df['G3'] >= 15]
if len(high_G3) > 0:
    bayes_prob = len(high_G3[high_G3['failures'] == 0]) / len(high_G3)
    print(f"5. Bayes Theorem:= {bayes_prob:.3f}")
else:
    print("5. No students with G3>=15")
```
