```python
restore_hotel = df[df['hotel'] == 'Resort Hotel']
restore_hotel['is_canceled'].value_counts(normalize= True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel' ]
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
restore_hotel = restore_hotel.groupby('reservation_status_date')[['adr']].mean()

city_hotel = city_hotel.groupby('reservation_status_date') [['adr']].mean()
```

```python

plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
```

```python



city_hotel_daily = city_hotel.groupby('reservation_status_date')['adr'].mean()
resort_hotel_daily = resort_hotel.groupby('reservation_status_date')['adr'].mean()

plt.figure(figsize=(20, 8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel_daily.index, resort_hotel_daily.values, label='Resort Hotel', linewidth=2)
plt.plot(city_hotel_daily.index, city_hotel_daily.values, label='City Hotel', linewidth=2)


```

```python

canceled_df = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Create barplot
sns.barplot(x='month', y='adr', data=canceled_df)

plt.xlabel('Month', fontsize=15)
plt.ylabel('Total Daily Rate', fontsize=15)
```
