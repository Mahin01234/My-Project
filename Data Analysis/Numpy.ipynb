```python
# Mean
import numpy as np
marks = np.array([70, 80, 90, 85, 75])
avg_marks = np.mean(marks)
print("Average Marks:", avg_marks)
```

```python
prices = np.array([100, 200, 300, 400, 5000])
median_price = np.median(prices)
print("Median Price:", median_price)
```

```python
marks = np.array([70, 80, 90, 85, 75])
std_marks = np.std(marks)
print("Standard Deviation:", std_marks)
```

```python
np.var(marks)
```

```python
var_marks = np.var(marks)
print("Variance of Marks:", var_marks)
```

```python
total_marks = np.sum(marks)
print("Total Marks:", total_marks)
```

```python
max_mark = np.max(marks)
print("Maximum Marks:", max_mark)
```

```python
min_mark = np.min(marks)
print("Minimum Marks:", min_mark)
```

```python
p50 = np.percentile(marks, 50)  # Median
p90 = np.percentile(marks, 90)  # Top 10% boundary
print("50th Percentile (Median):", p50)
print("90th Percentile:", p90)
```
