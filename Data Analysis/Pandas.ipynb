```python
import pandas as pd
import json
from io import StringIO

csv_content = """StudentID,FullName,Data Structure Marks,Algorithm Marks,Python Marks,CompletionStatus,EnrollmentDate,Instructor,Location
PH1001,Alif Rahman,85,85,88,Completed,2024-01-15,Mr. Karim,Dhaka
PH1002,Fatima Akhter,92,92,,In Progress,2024-01-20,Ms. Salma,Chattogram"""

# CSV to DataFrame
df = pd.read_csv(StringIO(csv_content))

# JSON to string
json_str = df.to_json(orient='records', indent=2)
print("JSON String:")
print(json_str)

# JSON file
with open('student_data.json', 'w', encoding='utf-8') as f:
    f.write(json_str)
```

```python
import json


students_data = {
    "metadata": {
        "total_students": 20,
        "data_source": "Student Database",
        "created_date": "2024-01-01"
    },
    "students": []
}


df = pd.read_csv('student_data (2).csv')

for _, row in df.iterrows():
    student_info = {
        "id": row['StudentID'],
        "name": row['FullName'],
        "marks": {
            "data_structure": row['Data Structure Marks'],
            "algorithm": row['Algorithm Marks'],
            "python": row['Python Marks']
        },
        "status": row['CompletionStatus'],
        "location": row['Location']
    }
    students_data["students"].append(student_info)

# JSON
with open('students_nested.json', 'w', encoding='utf-8') as f:
    json.dump(students_data, f, indent=2, default=str)

print("Nested JSON file created!")
```

```python
import json

with open('student_data.json', 'r') as file:
    students = json.load(file)


completed_students = [s for s in students if s['CompletionStatus'] == 'Completed']
print(f"Completed students: {len(completed_students)}")


python_high = [s for s in students if pd.notna(s.get('Python Marks')) and s['Python Marks'] >= 85]
print(f"Students with Python marks >= 85: {len(python_high)}")
```

```python
import json

def validate_json(file_path):

    try:
        with open(file_path, 'r') as file:
            json.load(file)
        print(f"{file_path} is valid JSON")
        return True
    except json.JSONDecodeError as e:
        print(f"Invalid JSON: {e}")
        return False

def prettify_json(input_file, output_file):
    with open(input_file, 'r') as f:
        data = json.load(f)

    with open(output_file, 'w') as f:
        json.dump(data, f, indent=2, sort_keys=True)

    print(f"Prettified JSON saved to {output_file}")

validate_json('student_data.json')
prettify_json('student_data.json', 'student_data_pretty.json')
```

**Apply Function**

```python
mn = df['Total Marks'].min()
mx = df['Total Marks'].max()


df['Scaled Marks'] = df['Total Marks'].apply(lambda x : (x-mn)/(mx-mn))


df
```

```python
# custom built function

def grading_system(marks):

    if marks >=260:
        return 'A+'
    elif marks>=250:
        return 'A'
    else:
        return 'A-'


df['Grade'] = df['Total Marks'].apply(grading_system)


df
```

```python
def marking_system(df):
    a = df['Data Structure Marks'] * 2
    b = df['Python Marks'] * 3
    c = df['Algorithm Marks'] * 4

    return a+b+c



df['Exceptional Marks'] = df.apply(marking_system,axis=1)


df
```
