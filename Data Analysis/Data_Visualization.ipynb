**Bar Chart**

```python

data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)

df
```

```python

# gender wise

gender_group = df.groupby('Gender').size()

gender_group.index
gender_group.values
```

```python

plt.bar(gender_group.index,gender_group.values,color=['orange','red'],edgecolor='black')

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender among students')
```

```python

#  Bar chart on real data

df = pd.read_csv('student_IQdata.csv')

df
```

```python
status = df.groupby('IQ_Score').size()

plt.bar(status.index,status.values,color=['green','orange','red'],edgecolor='black')

plt.xlabel('IQ Score')
plt.ylabel('Count')
plt.title('Bar Chart of IQ Score Frequency')
plt.show()
```

**pie chart**

```python
status = df.groupby('IQ_Score').size()
plt.pie(status, labels=status.index, autopct='%1.1f%%', shadow=True)
plt.title('Distribution of IQ Scores')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
student = pd.read_csv('sns_data.csv')
student.head()
```

**Lineplot using Seaborn**

```python

#axes level
sns.lineplot(data= student , x = 'week' , y = 'attendance_rate',errorbar=None)
```

```python
sns.lineplot(data= student , x = 'week' , y = 'attendance_rate',errorbar=None,hue='gender')
```

```python
sns.relplot(kind ='line', data= student , x = 'week' , y = 'test_score',errorbar=None,hue='class_level')
```

```python
sns.relplot(kind='scatter', data = student , x = 'study_hours' , y ='test_score',hue='gender',style='subject',size='Tshirt_size')
```
