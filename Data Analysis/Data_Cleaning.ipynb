**Ordinal Encoding**

```python
df = pd.DataFrame({"Size" : ["s" , "a" , "c" , "d" , "s" ]})
df
```

```python
df.sort_values(by='Size')
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_data = [["a" , "c" , "d" , "s" ]]
oe = OrdinalEncoder(categories = ord_data)
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
df
```

```python

df['Size_en'].plot(kind='line', figsize=(8, 4), title='Size_en')
```

```python

df.groupby('Size').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
```

```python

df['Size_en'].plot(kind='hist', bins=20, title='Size_en')
```

```python
ord_data = {"s" : 10 , "a" : 1 , "c" : 2 , "d" : 4 }

ord_data
```

```python
df["Size"].map(ord_data)
```

```python
df["Size_em_map"] = df["Size"].map(ord_data)
```

```python
df
```

```python
sns.boxplot (x = "Size_em_map" , data = df)
```

```python
sns.displot (df ["Size_em_map"])
```

```python
Q1 = df['Size_em_map'].quantile(0.25)
Q3 = df['Size_em_map'].quantile(0.75)
```

```python
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
lower_bound , upper_bound
```

```python
import numpy as np
df = df[(df['Size_em_map'] >= lower_bound) & (df['Size_em_map'] <= upper_bound)]

df['Size_em_map'] = np.where(df['Size_em_map'] > upper_bound, upper_bound,
                       np.where(df['Size_em_map'] < lower_bound, lower_bound,
                                df['Size_em_map']))
```
